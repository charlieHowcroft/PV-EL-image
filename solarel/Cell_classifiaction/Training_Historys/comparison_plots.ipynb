```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
history_efficientnet_v2_L_df = pd.read_csv("training_history_focal.csv")
plt.plot(history_efficientnet_v2_L_df['accuracy'])
plt.plot(history_efficientnet_v2_L_df['val_accuracy'])
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Training', 'Validation'], loc='upper left')
plt.ylim(0, 1)
plt.show()
```

```python
history_efficientnet_v2_L_df = pd.read_csv("training_history.csv")
plt.plot(history_efficientnet_v2_L_df['accuracy'])
plt.plot(history_efficientnet_v2_L_df['val_accuracy'])
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Training', 'Validation'], loc='lower right')
plt.ylim(0, 1)
plt.show()
```

```python
# List of training history file names
history_files = ["training_history.csv", "training_history_focal.csv"]
names = [""]

# Define colors for training and validation curves
color = ['blue', 'red']

# Plot the training and validation accuracy curves
for i, file_name in enumerate(history_files):
    # Read the CSV file into a pandas DataFrame
    history_df = pd.read_csv(file_name)
    
    # Plot training accuracy curve
    plt.plot(history_df['accuracy'], color=color[i], label='Training ' + str(i+1))
    
    # Plot validation accuracy curve with transparency
    plt.plot(history_df['val_accuracy'], color=color[i], alpha=0.7, label='Validation ' + str(i+1))

plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.ylim(0.5, 1)
plt.grid(True)
plt.show()
```
