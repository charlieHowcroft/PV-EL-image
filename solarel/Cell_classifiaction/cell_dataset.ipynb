```python
import cv2
import os
import pandas as pd
```

rename existing dataset

```python
def rename_images(folder_path):
    image_count = 883

    for filename in os.listdir(folder_path):
        image_path = os.path.join(folder_path, filename)
        new_filename = f"image_{image_count}.jpg"
        new_image_path = os.path.join(folder_path, new_filename)

        os.rename(image_path, new_image_path)
        print(f"Renamed '{filename}' to '{new_filename}'")
            
        image_count += 1
```

```python
# Usage example
folder_path = "C:/Users/chuck/OneDrive/Desktop/Honors/elpv-dataset/images"
rename_images(folder_path)
```

extend csv file

```python
csv_path = "C:/Users/chuck/OneDrive/Desktop/filtered_df.csv"
df = pd.read_csv(csv_path)
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
print(df)
```

```python
# Define the range of new rows to be added
start_index = 0  # Starting index
end_index = 749  # Ending index (exclusive)

# Iterate over the range and add new rows
for i in range(start_index, end_index):
    filename = f"image_{i}.jpg"
    new_row = pd.DataFrame({
        'image_name': [filename],
        'good': [0.0],
        'crack': [0.0],
        'corrossion': [0.0]
    })
    df = df.append(new_row, ignore_index=True)

# Print the updated DataFrame
print(df)
```
