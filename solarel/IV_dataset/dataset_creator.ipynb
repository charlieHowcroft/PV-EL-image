```python
import pandas as pd
import cv2 as cv
import os
```

```python
folders = ["C:/Users/chuck/OneDrive/Desktop/IV_curve_prediction/Images/BT1", "C:/Users/chuck/OneDrive/Desktop/IV_curve_prediction/Images/M0060"]

def get_image_names(folders):

    files = []
    for folder in folders:
        for dirpath, _, filenames in os.walk(folder, topdown=False):
            for filename in filenames:
                print(type(filename))
                files.append(filename[0:-4])
    return files
```

```python
image_names = get_image_names(folders)
print(image_names)
```

```python
new_names = []
for name in image_names:
    new_names.append(name[:-9])
```

```python
iv_curve_folder = "C:/Users/chuck/OneDrive/Desktop/IV_curve_prediction/I-V_data"
def delete_csv_files_with_0001(folder_path):
    for root, dirs, files in os.walk(folder_path):
        for filename in files:
            if filename.endswith('.csv') and '_0001' in filename:
                file_path = os.path.join(root, filename)
                os.remove(file_path)
                print(f"Deleted: {file_path}")

delete_csv_files_with_0001(iv_curve_folder)
```

```python

def check_csv_files(folder_path, allowed_names):
    for root, dirs, files in os.walk(folder_path):
        for filename in files:
            if filename.endswith('.csv') and filename not in allowed_names:
                file_path = os.path.join(root, filename)
                os.remove(file_path)
                print(f"removing: {filename}")


check_csv_files(iv_curve_folder, new_names)
```
